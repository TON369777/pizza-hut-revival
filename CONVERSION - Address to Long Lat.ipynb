```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AllPizzaStore_FORMATTED2.csv')
```

```python
df
```

```python
l_cols_concat = ['City', 'State', 'Zip', 'Country']
```

```python
df['unique_address'] = df['Street'].str.cat(others=df[l_cols_concat], sep=',', na_rep='')
```

```python
df
```

```python
address1 = df['unique_address'].iloc[0]
address1
```

```python
from geopy.geocoders import ArcGIS
```

```python
geolocator_arcgis= ArcGIS()
location = geolocator_arcgis.geocode(address1)
print('Latitute: '+str(location.latitude)+', Longitude: '+str(location.longitude))
```

```python
def service_geocode(g_locator, address):
    location = g_locator.geocode(address)
    if location!=None:
        return(location.latitude, location.longitude)
        print('Latitute: '+str(location.latitude)+', Longitude: '+str(location.longitude))
    else:
        return np.NaN
```

```python
def service_geocode_lat(g_locator, address):
    location = g_locator.geocode(address)
    if location!=None:
        return(location.latitude)
    else:
        return np.NaN
    
def service_geocode_long(g_locator, address):
    location = g_locator.geocode(address)
    if location!=None:
        return(location.longitude)
    else:
        return np.NaN
```

```python
df['LAT_arcgis'] = df['unique_address'].apply(lambda x: service_geocode_lat(geolocator_arcgis, x))
print("Conversion Complete")
```

```python
df['LON_arcgis'] = df['unique_address'].apply(lambda x: service_geocode_long(geolocator_arcgis, x))
print("Conversion Complete")
```

```python
df['LAT_LON_arcgis'] = df['unique_address'].apply(lambda x: service_geocode(geolocator_arcgis, x))
print("Conversion Complete")
```
